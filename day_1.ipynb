```python
import pandas as pd

day_01 = pd.read_csv('inputs/day_01.txt', sep="\n",names=['depths'])
day_01.head()
```

## Part 1

```python
larger = 0
depth_pre = 0
for idx, depth in day_01.iterrows():
    depth_post = depth[0]
    comp = depth_post - depth_pre > 0
    if comp and depth_pre != 0:
        larger += 1
    depth_pre = depth_post

larger
```

## Part 2

```python
window_larger = 0
comps = day_01.shape[0]
comp_pre = day_01.iloc[0:3,:].sum()[0]

for i in range(4, comps+1):
    comp_post = day_01.iloc[i-3:i,:].sum()[0]
    print("i, comp_post:",i,comp_post)
    comp_vals = comp_post - comp_pre > 0
    if comp_vals:
        window_larger += 1
    comp_pre = comp_post
window_larger
```
